# Data Cleaning

```python
#rename column name 'No_Pation' to Patient_No
dfcopy=dfcopy.rename(columns={"No_Pation":"Patient_No"})
print(dfcopy.to_string())
```

# Remove or fill missing values

```python
#check null values
dfcopy.isnull().sum()
```

```python
#fill values in age column
new_age=df["AGE"].median()
new_age
```

```python
dfcopy["AGE"]=dfcopy["AGE"].fillna(new_age)
print(dfcopy.to_string())
```

```python
#fill values in HbA1c column
new_HbA1c=df["HbA1c"].mean()
new_HbA1c
```

```python
dfcopy["HbA1c"]=df["HbA1c"].fillna(new_HbA1c)
print(dfcopy.to_string())
```

```python
#drop remaining missing values
dfcopy=dfcopy.dropna()
print(dfcopy.to_string())
```

```python
dfcopy.isnull().sum()
```

# Graphs

# Histogram for AGE:

```python
import matplotlib.pyplot as plt
plt.hist(dfcopy['AGE'],bins=20,color='purple',edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# Bar Chart for Gender:

```python
gender_counts=dfcopy['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color='purple')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
