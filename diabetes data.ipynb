# data cleaning

```python
#rename column name No_Pation to Patient_No
df.rename(columns={"No_Pation":"Patient_No"},inplace=True)
print(df.to_string())
```

# remove or fill missing values

# check null values

```python
df.isnull().sum()
```

```python
#fill missing values in column age
new_value=df["AGE"].median()
new_value
```

```python
df["AGE"].fillna(new_value,inplace=True)
print(df.to_string())
```

```python
#fill missing values in column  HbA1c
mean_value=df["HbA1c"].mean()
mean_value
```

```python
df["HbA1c"].fillna(mean_value,inplace=True)
print(df.to_string())
```

```python
#to confirm the change
df.isnull().sum()
```

```python
#Dropping the missing values of other columns
df.dropna(inplace=True)
df.isnull().sum()
```
